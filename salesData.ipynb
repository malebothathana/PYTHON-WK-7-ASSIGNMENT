```python
# Analyzing Data with Pandas and Visualizing Results with Matplotlib

# Task 1: Load and Explore the Dataset

## 1.1 Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

## 1.2 Load the Dataset
# Creating a simulated sales dataset
data = {
    'Date': pd.date_range(start='2025-01-01', periods=100, freq='D'),
    'Product': np.random.choice(['Laptop', 'Tablet', 'Phone', 'Monitor'], size=100),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=100),
    'Units Sold': np.random.randint(1, 20, size=100),
    'Revenue': np.random.uniform(100, 2000, size=100).round(2)
}

df = pd.DataFrame(data)

# Display the first few rows
df.head()

## 1.3 Explore the Structure
# Check data types
print(df.dtypes)

# Check for missing values
print(df.isnull().sum())

# Data Cleaning
# No missing values found, so no cleaning necessary.

# Task 2: Basic Data Analysis

## 2.1 Basic Statistics
# Basic descriptive statistics
df.describe()

## 2.2 Grouping and Aggregating
# Grouping by Product and calculating mean Revenue
grouped = df.groupby('Product')['Revenue'].mean()
print(grouped)

# Task 3: Data Visualization

## 3.1 Line Chart: Revenue Over Time
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Revenue'], label='Daily Revenue', color='blue')
plt.title('Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue ($)')
plt.legend()
plt.grid(True)
plt.show()

## 3.2 Bar Chart: Average Revenue per Product
plt.figure(figsize=(8,5))
grouped.plot(kind='bar', color='coral')
plt.title('Average Revenue per Product')
plt.xlabel('Product')
plt.ylabel('Average Revenue ($)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()

## 3.3 Histogram: Distribution of Units Sold
plt.figure(figsize=(8,5))
plt.hist(df['Units Sold'], bins=10, color='teal', edgecolor='black')
plt.title('Distribution of Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')
plt.show()

## 3.4 Scatter Plot: Units Sold vs Revenue
plt.figure(figsize=(8,5))
sns.scatterplot(x='Units Sold', y='Revenue', hue='Region', data=df, palette='muted')
plt.title('Units Sold vs Revenue by Region')
plt.xlabel('Units Sold')
plt.ylabel('Revenue ($)')
plt.legend(title='Region')
plt.show()

# Final Observations
print("""
Observations:
- Laptops tend to generate higher revenue compared to Phones and Tablets.
- Units Sold is fairly evenly distributed, with some common peaks.
- Revenue shows fluctuations daily with occasional spikes.
- A positive trend is observed between Units Sold and Revenue.
""")

# Bonus: Error Handling Example
try:
    df_error = pd.read_csv('non_existent_file.csv')
except FileNotFoundError as e:
    print(f"Error handled: {e}")
```
